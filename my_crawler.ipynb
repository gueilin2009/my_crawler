```python
#Request要爬取的網頁
import requests

html = requests.get('https://tw.appledaily.com/charity/projlist/')    #點選"我要捐款"小圖示,進入"捐款進度報告"主頁
```

```python
html.text   #encoding為文字格式
```

```python
#為parse做準備
from bs4 import BeautifulSoup
import re  #regular expression,兩者連用加上compile

soup = BeautifulSoup(html.text, "lxml")
soup
```

# step 2  (1)

```python
#找出所有tr的tag,會是一個list,每個tr下包td
all_rows = soup.find_all(re.compile("tr"))  #或直接tr
all_rows[:3]  #查看前三列,第一列是表格標題,即第一個tr下
```

```python
#把標題存成column_name變數
column_name = all_rows[0]  #取出list中的第一個項目,是bs4.element.Tag()
#print(type(all_rows[0]))
column_name
```

```python
column_name = [text for text in column_name.stripped_strings]   #斷開字串,存回list以便建立dataframe
#column_name = [text for text in column_name]
column_name
```

```python
#創一個dataframe,存入標題
import pandas as pd
table = pd.DataFrame(columns = column_name)
table
```

```python
#從第二個tr開始找資料    type(all_rows[1:])=list

for i, row in enumerate(all_rows[1:]):  #i為設定將資料一筆一筆存入
    data = [s for s in row.stripped_strings]   #每個row都是bs4.element.Tag()
    print(data)  #印出全部資料
    #table.loc[i]=data   #Error!欄位沒有剛剛好,第一筆資料沒有累計(元)
```

```python
#將資料存入dataframe,並輸出檔案 (從第二筆開始)
!pip install xlrd

for i, row in enumerate(all_rows[2:]): 
    data = [s for s in row.stripped_strings] 
    table.loc[i]=data
table
#table.to_excel("./my_crawler/report_1.xlsx")
```

# step 2 (2) 

```python
#另一種解法：找出所有td的tag,一次填入表格中

all_cells = soup.find_all(re.compile("td"))
all_cells[:12]  #查看前兩列,共12個td,第一列為標題
```

```python
#把td tag中所包的資訊存成list
list = []
for s in all_cells[12:]:   
    data = s.text
    list.append(data)
list
#type(list)
#len(list)  #共有114個td
```

```python
#創一個dataframe,存入上述資料,並輸出檔案
import pandas as pd
import numpy as np

matrix=[]   #創空矩陣, dataframe不會自動根據欄位數把list切份，先存入矩陣中,預設為直的填入
for i in range(0,len(list),6):
    matrix.append(list[i:i+6])
matrix    

matrix=np.array(matrix)  #將Matrix存成np.array後轉置
matrixT=matrix.T
matrixT

tables=pd.DataFrame()  #創空dataframe,將matrix存入
for i in range(0,len(matrixT)):
    tables['%s'%i]=matrixT[i]
tables

tables = tables.rename(columns={'0':'編號','1':'報導標題','2':'刊登日期','3':'狀態','4':'累計(元)','5':'捐款明細'})
tables
#tables.to_excel("./my_crawler/report_2.xlsx")
```

# step 3

```python
#抓取"報導標題"欄位的url連結,增加到原來的dataframe中
import pandas as pd

article_url = [tag.find("a")['href'] for tag in soup.find_all("h2")[3:]]   #第三個"h2"開始下面都有a
article_url  #type為list
```

```python
#將list存成Series與DataFrame
#text_url = pd.Series()
#text_url = pd.Series(article_url)  #存成Series
#text_url.append(text_url)

table['text_url'] = pd.Series(article_url)  #存入dataframe並對series命名
table.head() #只顯示前五筆
```

```python
#抓取"捐款明細"欄位的url連結,增加到原來的dataframe中

web_url = 'https://tw.appledaily.com/'     #原來的href="/charity/projdetail/proj/[A-Z]+[0-9]"是一個字串,加上web_url
detail_url = [web_url + tag['href'] for tag in soup.find_all("a", {"class":"details"})]  #在href在class:details下
detail_url
detail_url.pop(0)  #刪去第一筆
detail_url
```

```python
table['detail_url'] = pd.Series(detail_url) 
table.head()
#table.to_excel("./my_crawler/detail_url.xlsx")
```

# step 4

```python
#抓取全部的網頁數
page = int(soup.find("div", {"class":"ui-pagination align-center"}).find_all("a")[-3].text)   #最下面的bar
page
```

```python
#隨機Request其中一頁
import requests
import sys

for i in range(1, page):
    sys.stdout.write('正在抓取捐款進度報告第'+ '%s'%i +'頁')
    craw = requests.get('https://tw.appledaily.com/charity/projlist/Page/%s'%i)
    soup = BeautifulSoup(craw.text, 'lxml')
```

# step 5

```python
#抓取其中一頁其中一筆已結案者的捐款明細資訊  
import requests
from bs4 import BeautifulSoup

closed = requests.get('https://tw.appledaily.com/charity/projdetail/proj/A0050')
soup = BeautifulSoup(closed.text, 'lxml')
donate = soup.find_all("table")[1].find_all("td")   #從第二個table tag開始抓,"class":"donabox-detail"
donate  #方法同上
```

```python
list = []
for s in donate: 
    data = s.text
    list.append(data)
list
```

```python
import pandas as pd
import numpy as np

matrix=[] 
for i in range(0,len(donate),4):
    matrix.append(list[i:i+4])
    
matrix=np.array(matrix)  
matrixT=matrix.T

donate_table=pd.DataFrame()
for i in range(0,len(matrixT)):
    donate_table['%s'%i]=matrixT[i]

donate_table = donate_table.rename(columns={'0':'筆數','1':'捐款人姓名','2':'捐款金額(元)','3':'捐款日期',})
donate_table.tail(30)  #只顯示後30筆
#donate_table.to_excel("./my_crawler/donate.xlsx")
```

```python
#爬取主網頁的文章(含文章代號,主副標題,內文)

import requests 
from bs4 import BeautifulSoup

#Request主網頁
url = 'https://tw.appledaily.com/headline/daily/20160704/37293645'  
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")

#爬取主副標題
main_title = soup.find("article", {"class":"ndArticle_leftColumn"}).find("h1").text   
sub_title = soup.find("article", {"class":"ndArticle_leftColumn"}).find("h2").text 
print(main_title, sub_title)

#爬取全文 
paragraph = [tag.text for tag in soup.find("div", {"class":"ndArticle_margin"}).find_all("p")]   #特定的class
print(paragraph)    #後面不在文章內的廣告說明可用tag p[2],[3]斷開
#type(paragraph)
```
