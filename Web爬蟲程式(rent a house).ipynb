# Rent a house 

```python
import requests   
import json

#上591租屋網 https://rent.591.com.tw, 搜尋'臺南市'
url = "https://rent.591.com.tw/home/search/rsList" #點Network找到rsList, url=https://rent.591.com.tw/home/search/rsList?is_new_list=1&type=1&kind=0&searchtype=1&region=15

#data
headers = json.loads(r'''{
                    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36"
                    }''')

qs = {"is_new_list": "1",
      "type": "1",
      "kind": "1",
      "searchtype": "1",
      "region": "15",
      "firstRow": "1"  # pagination, increment this to set the OFFSET
      }

res = requests.get(url, headers=headers, params=qs)
res.json() #直接轉json格式
```

```python
# Save dict to JSON file   把.json格式倒到檔案中
import json

json.dump(res.json(), open('./rent/591.json', 'w'), ensure_ascii=False)  #ascii碼為F
```

```python
#讀出來看看  #Linux系統直接 head ./rent/591.json 

with open('./rent/591.json') as f:
    rent = f.read()
```
