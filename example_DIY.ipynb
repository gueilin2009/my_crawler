# DIY 1 : 博客來網路書店 GET  

```python
#Request網頁
import requests

response = requests.get("http://search.books.com.tw/search/query/key/python/cat/all")   #搜尋關鍵字"python"後  
response.text
```

```python
#為parse做準備
from bs4 import BeautifulSoup 

soup = BeautifulSoup(response.text, "lxml")  #以lxml解析
soup.title.string
```

```python
#爬取書名
titles = [tag["alt"] for tag in soup.find_all("img", {"class":"itemcov"})]   #要加上class爬下來才會是第一頁的
titles
```

```python
#存入欄位
import pandas as pd

pd.Series(titles)  #出現index
#books = pd.Series()  #起Series欄位名稱
df = pd.DataFrame({'書名':pd.Series(titles)})  #直接加df欄位名稱
df
```

```python
#爬取價格
price = [tag.text for tag in soup.find_all("strong")]  #價格欄的bar下有span class=price tag,但沒有全部都有書籍的strong tag
price.pop(0)    #取出第一項 '1'
price.pop(-1)   #取出最後一項 '全部分類'     #list out of range時重定位tag
price

dollars= [] 
for i in range(1,len(price),2):   #取出第二個,即價格
    dollars.append(price[i])
dollars
```

```python
#新增入表格
df['價格'] = pd.Series(dollars)
df
```

# DIY 2  : 台鐵火車時刻表  POST+JSON

```python
#Request網頁
import requests

data = {                        #爬取2019/01/01臺南往臺北所有的車班 Form Data
    "FromStationName": "1228",
    "ToStationName": "1008",
    "TrainClass": "2",
    "searchdate": "2019-01-01",
    "FromTimeSelect": "0000",
    "ToTimeSelect": "2359",
    "Timetype": "1"   
}

res = requests.post ("http://twtraffic.tra.gov.tw/twrail/TW_SearchResult.aspx", data=data)  #request為POST
res.text 
```

```python
#為pase做準備
from bs4 import BeautifulSoup 

soup = BeautifulSoup(res.text, "lxml")  #lxml解析
soup   
```

```python
#取出JSON資料
train_data = [tag.text for tag in soup.find_all("script") if 'JSONData' in tag.text]   #車班資料在script中,先找出所有script tag,再用if關鍵字in找比較快, 開發人員工具Newwork下的response也可以看
train_data  #是list
train = train_data[0]  #取出來後是str,再刪掉前面不需要的字
```

```python
train
```

```python
#取出JSON資料
train = train[13:-1]   #最後一個字不要
train   #是text
```

```python
#Text轉成Json格式
import json
trains = json.loads(train)    #list type
trains   
```

```python
print(type(trains), len(trains))   #共有28筆資料,此JSON資料結構的名稱為trains
```

```python
#一次存入dataframe
import pandas as pd

df1 = pd.DataFrame(trains)   #List可直接import進dataframe
df1.to_excel("./trains.xlsx", encoding = 'big5')  #輸出檔案再選取所需欄位(與訂票系統畫面不完全相同)
df1
```

```python
#以JSON為資料結構只撈取使用者需要的資料

#表格名稱為trains
trains[0]  #第一筆資料
trains[-1]  #最後一筆資料
type(trains[0])  #每一筆資料皆為一筆JSON

df2 = pd.DataFrame()  
df2['車種']=[columns['Class_Code'] for columns in trains]
df2['車次']=[columns['Train_Code'] for columns in trains]
df2['發車站']=[columns['Begin_EName'] for columns in trains]
df2['終點站']=[columns['End_EName'] for columns in trains]
df2['開車時間']=[columns['From_Departure_Time'] for columns in trains]
df2['到達時間']=[columns['To_Arrival_Time'] for columns in trains]
df2['備註']=[columns['Comment'] for columns in trains]
df2['票價']=[columns['Fare'] for columns in trains]
df2.head()
```

# DIY 3 : 文化部展覽活動  JSON+JSON

```python
#Request網頁
import requests  
                                  #網址直接給定JSON格式, JSON包JSON
show = requests.get("http://cloud.culture.tw/frontsite/trans/SearchShowAction.do?method=doFindTypeJ&category=6")  
#print(show.text)   
```

```python
#轉成JSON格式
import json

showj =json.loads(show.text)
showj
```

```python
#查看總筆數
print(type(showj), len(showj))    #共650筆JSON資料,每日更新
```

```python
#存入dataframe
import pandas as pd

df3 = pd.DataFrame(showj)
df3
```

```python
#顯示showInfo欄位的資訊

df3['showInfo'][0]
#type(df3['showInfo'][0])  #每細格皆是一個list,包json資料,可直接import進dataframe
```

```python
#將showInfo的JSON資料存到另一個dataframe中
import pandas as pd

pd.DataFrame(df3['showInfo'][0])   #先存第一筆
```

```python
df4 = pd.DataFrame()
for i in df3['showInfo']:
    df4 = df4.append(pd.DataFrame(i)).reset_index(drop=True)   #創一個空的df,append新資料
df4
```

```python
#將上一個dataframe所需的資料加到此dataframe

df4['title']=df3['title']
df4
```
